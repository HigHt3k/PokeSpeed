```python
import csv
import pokebase as pb


def get_pokemon_stats(pokemon):
    stats = [pokemon.stats[i].base_stat for i in range(0, 6)]
    return stats

def load_all():
    poke_list = []
    for id in range(1, 920):
        try:
            poke_list.append(pb.pokemon(id))
            print(poke_list[id-1])
        except:
            pass
    return poke_list

def load_one(id):
    return pb.pokemon(id)

def save_to_csv(pokemons):
    with open('poke_stats', 'w') as f:
        # using csv.writer method from CSV package
        write = csv.writer(f)

        write.writerow(["id","Name","HP","ATK","DEF","SPATK","SPDEF","SPEED"])

        for pokemon in pokemons:
            write.writerow([pokemon.id,
                            pokemon,
                            get_pokemon_stats(pokemon)[0],
                            get_pokemon_stats(pokemon)[1],
                            get_pokemon_stats(pokemon)[2],
                            get_pokemon_stats(pokemon)[3],
                            get_pokemon_stats(pokemon)[4],
                            get_pokemon_stats(pokemon)[5]])

poke_list = load_all()
save_to_csv(poke_list)
```
